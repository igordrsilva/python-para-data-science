## 10 Principals Methods in Python to Initial Data Analysis

The dataset we'll use in this example is available on Kaggle: [Gym Membership Dataset](https://www.kaggle.com/datasets/ka66ledata/gym-membership-dataset)

### ⤵️ Imports

```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("ka66ledata/gym-membership-dataset")
```

### 💡 Starting with Data Analysis

#### 1. `pd.DataFrame().head()`
It shows the five first data's lines. Here we can have an idea of what kind of data we'll work.

```python
data = pd.read_csv(os.path.join(path, "gym_membership.csv"))
data.head()
```

**🗣️ An extra tip:** `pd.DataFrame().tail()`

Returns the five last data's lines.

```python
data.tail()
```

#### 2. `pd.DataFrame().shape`
Returns the dataset dimention (lines, columns)

```python
data.shape
```

#### 3. `pd.DataFrame().info()`
Shows the structural resume: columns name, count of null values and types of data.

```python
data.info()
```

#### 4. `pd.DataFrame().describe()`
Shows main descriptives statistics (returns numerics data by default).

```python
data.describe()
```

**🗣️ An extra tip:** `pd.DataFrame().describe(include='object')`

Returns some statistics of categorical data.

```python
data.describe(include='object')
```

#### 5. `pd.DataFrame()[column].nunique()`
Returns the count of unique values in a column.

```python
data['fav_group_lesson'].nunique()
```

#### 6. `pd.DataFrame()[column].value_counts()`
Returns the frequency of each value in a column.

```python
data['fav_group_lesson'].value_counts()
```

#### 7. `pd.DataFrame()[column].mean()`
Returns the mean of column value.

```python
data['visit_per_week'].mean()
```

**🗣️ An extra tip:** `pd.DataFrame[column].mode()` e `pd.DataFrame[column].median()`

More of central tendency statistics. Returns the mode and median of column value.

```python
data['visit_per_week'].mode()
```

```python
data['visit_per_week'].median()
```

#### 8. `pd.DataFrame()[column].std()`
Calculate the standard deviation.

```python
data['visit_per_week'].std()
```

#### 9. `pd.DataFrame().dtypes`
Returns the columns types of dataset.

```python
data.dtypes
```

#### 10. `pd.DataFrame().columns`
Returns all columns names of dataset.

```python
data.columns
```
